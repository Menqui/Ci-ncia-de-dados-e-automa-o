```python
#selenium : uma biblioteca no python para automação web
#selenium controla o  navegador, diferente do pyautogui
#necessário o webdriver para controlar o navegador
#se for google chrome(chromedriver) se for firefox(geckodriver)
```

```python
#criar um navegador para controlar o selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time


#automaticamente indentifica a sua versão atual do google e baixa o webdriver compativel com a mesma
#servico = Service(ChromeDriverManager().install())

navegador = webdriver.Chrome()
```

```python
#abre o navegador e conecta ao youtube
navegador.get('https://www.google.com')
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('shinenoga e waa lyrics')
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

#uzar um scroll para descer a tela , referencia dos paramentros em x e y
navegador.execute_script("window.scrollTo(0,1570);")
#entra no video do youtube
navegador.find_element('xpath','//*[@id="kp-wp-tab-default_tab:kc:/music/recording_cluster:lyrics"]/div[2]/div/div/div/div/div[2]/div/div[2]/div[2]/div[1]/div/div/video-voyager/div/div[1]/div/a/div/div[1]/div/span').click()
```

```python
#pyautogui também é uma biblioteca para automação porém ele não controla o navegador em sim e sim o computador
#por isso não pode ser executado em segundo plano

import pyautogui
```

```python
#mesma atividade (entrar em um vídeo do youtube) porém usando pyautogui
pyautogui.PAUSE = 1
# pyautogui.click -> clicar
# pyautogui.press -> apertar 1 tecla
# pyautogui.hotkey -> conjunto de teclas (ex : alt + tab)
# pyautogui.write -> escreve um texto

#controla os atalhos do teclado
pyautogui.hotkey('ctrl','t')
pyautogui.write('shinenoga e wa')
pyautogui.press('enter')
pyautogui.click(x=535,y=549)#clica na aba do video
```

```python
#pega a posição exata dos pixel na tela onde esta o mouse
time.sleep(5)#espera 5 segundos antes de executar a ação
pyautogui.position()
```
